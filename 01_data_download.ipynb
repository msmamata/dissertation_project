```python
#---------2- Downloading Brent Crude Oil Price---------
from fredapi import Fred
import pandas as pd
from dotenv import load_dotenv
import os

# Load .env file
load_dotenv()

# Get API key from environment
FRED_KEY = os.getenv("FRED_API_KEY")
if not FRED_KEY:
    raise ValueError("API key not found. Check your .env file.")

fred = Fred(api_key=FRED_KEY)

print("Downloading Brent Crude from FRED...")

brent = fred.get_series(
    'DCOILBRENTEU',
    observation_start='1990-01-01',
    observation_end='2025-12-31'
)

df_brent = brent.reset_index()
df_brent.columns = ['date', 'brent_crude_usd']
df_brent.to_csv("dissertation_data/raw/brent_crude_raw.csv", index=False)

print(f"Done!")
print(f"Shape: {df_brent.shape}")
print(f"Date range: {df_brent['date'].min()} to {df_brent['date'].max()}")
print(df_brent.tail(5))
```

```python
#----cleaning the data, since investing.com files always need to do cleaning because of the format.
# Load raw fil
df_freight = pd.read_csv("dissertation_data/raw/freight_rate_raw.csv")

# Keep only Date and Price
df_freight = df_freight[['Date', 'Price']]

# Rename columns to standard format
df_freight.columns = ['date', 'baltic_dry_index']

# Remove commas from numbers e.g. 1,877.00 → 1877.00
df_freight['baltic_dry_index'] = (df_freight['baltic_dry_index']
                                   .astype(str)
                                   .str.replace(',', '')
                                   .astype(float))

# Convert date to standard datetime format
df_freight['date'] = pd.to_datetime(df_freight['date'], format='%m/%d/%Y')

# Filter to 1990 onwards
df_freight = df_freight[df_freight['date'] >= '1990-01-01']

# Sort oldest to newest (Investing.com gives newest first)
df_freight = df_freight.sort_values('date').reset_index(drop=True)

# Save clean version
df_freight.to_csv("dissertation_data/raw/baltic_dry_index_raw", index=False)

print(f"Done! Shape: {df_freight.shape}")
print(f"Range: {df_freight['date'].min()} to {df_freight['date'].max()}")
print(df_freight.head(5))
print(df_freight.tail(5))

# droped Open, High, Low, Vol, Change%

# Cleaned = ready for merging
df_freight.to_csv(
    "dissertation_data/processed/baltic_dry_index_clean.csv", 
    index=False
)
```

```python
#clean data

df_gpr = pd.read_excel("dissertation_data/raw/gpr_raw.xls")

# Columns to keep — global GPR + 6 countries
# Note: Bangladesh, Pakistan, Singapore not in GPR country list
# so we are using global GPR for those country, ####can also test other alternative ways to fill that void
keep_cols = [
    'month',
    'GPR',      # Global GPR index — main variable
    'GPRT',     # Threat component
    'GPRA',     # Act component
    'GPRC_IND', # India
    'GPRC_JPN', # Japan
    'GPRC_KOR', # South Korea
]

df_gpr_clean = df_gpr[keep_cols].copy()

# Rename for clarity
df_gpr_clean.columns = [
    'date',
    'gpr_global',
    'gpr_threat',
    'gpr_act',
    'gpr_india',
    'gpr_japan',
    'gpr_korea',
]

# Convert date to datetime
df_gpr_clean['date'] = pd.to_datetime(df_gpr_clean['date'])

# Filter to 1990 onwards
df_gpr_clean = df_gpr_clean[df_gpr_clean['date'] >= '1990-01-01']

# Sort oldest to newest
df_gpr_clean = df_gpr_clean.sort_values('date').reset_index(drop=True)

# Saving raw as CSV instead of xls, since newer versions of pandas dropped support for writing .xls format
df_gpr.to_csv("dissertation_data/raw/gpr_raw.csv", index=False)

# Save clean version
df_gpr_clean.to_csv(
    "dissertation_data/processed/gpr_clean.csv", 
    index=False
)

print(f"Done!")
print(f"Shape: {df_gpr_clean.shape}")
print(f"Range: {df_gpr_clean['date'].min()} to {df_gpr_clean['date'].max()}")
print(df_gpr_clean.head(5))
print(df_gpr_clean.tail(5))


##analyiss of the data. In feb the gpr global jumps to 116 and India spikes to 0.43, which is the war signal right there in the data
# for EDA: gpr_global:  116.7  ← significant spike
#gpr_india:   0.43   ← highest value in the entire dataset
#gpr_japan:   0.17   
#gpr_korea:   0.21
# when doing the EDA i can highlight this exact moment as validation that my GPR variable correctly captures the conflict event.
```

```python


df_res = pd.read_csv("dissertation_data/raw/oil_reserves_raw.csv")

# Fix column names to lowercase
df_res.columns = df_res.columns.str.strip().str.lower()

print("Fixed columns:", df_res.columns.tolist())

# Create full year range for each country
countries = df_res['country'].unique()
all_years = range(1990, 2026)

print(f"Countries: {countries}")
print(f"Years: {min(all_years)} to {max(all_years)}")

dfs = []
for country in countries:
    # Get data for this country
    df_c = df_res[df_res['country'] == country].copy()
    df_c = df_c.sort_values('year')

    # Create full year index
    df_full = pd.DataFrame({'year': all_years})
    df_full['country'] = country

    # Merge with known values
    df_full = df_full.merge(
        df_c[['year', 'reserve_days']],
        on='year',
        how='left'
    )

    # Interpolate missing years linearly
    df_full['reserve_days'] = (df_full['reserve_days']
                                .interpolate(method='linear')
                                .round(1))

    dfs.append(df_full)

# Combine all countries
df_reserves_full = pd.concat(dfs, ignore_index=True)
df_reserves_full['date'] = pd.to_datetime(
    df_reserves_full['year'].astype(str)
)
# Save
df_reserves_full.to_csv(
    "dissertation_data/raw/oil_reserves_raw.csv",
    index=False
)

print(f"\nDone! Shape: {df_reserves_full.shape}")
print(f"Countries: {df_reserves_full['country'].unique()}")
print(f"\nSample — Japan reserve days:")
japan = df_reserves_full[df_reserves_full['country'] == 'Japan']
print(japan[['year','reserve_days']].to_string(index=False))
```
